```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df =pd.read_csv("USA_Housing.csv")
df.head()
```

```python
df.rename(columns ={'Avg. Area Income':'Avg_Area_Income'}, inplace=True)
df.rename(columns ={'Avg. Area House Age':'Avg_Area_House_Age'}, inplace=True)
df.rename(columns ={'Avg. Area Number of Rooms':'Avg_Area_Number_of_Rooms'}, inplace=True)
df.rename(columns ={'Avg. Area Number of Bedrooms':'Avg_Area_Number_of_Bedrooms'}, inplace=True)
df.rename(columns ={'Area Population':'Area_Population'}, inplace=True)
```

```python
df.head()
```

```python
plt.scatter(df.Avg_Area_Income,df.Price,color='blue')
plt.xlabel("Avg Area Income")
plt.ylabel("Price")
plt.show()
```

```python
plt.scatter(df.Avg_Area_House_Age,df.Price,color='blue')
plt.xlabel("Avg Area House Age")
plt.ylabel("Price")
plt.show()
```

```python
plt.scatter(df.Avg_Area_Number_of_Rooms,df.Price,color='blue')
plt.xlabel("Avg Area Number of Rooms")
plt.ylabel("Price")
plt.show()
```

```python
plt.scatter(df.Area_Population,df.Price,color='blue')
plt.xlabel("Area_Population")
plt.ylabel("Price")
plt.show()
```

```python
df.isnull().sum()
```

```python
df.drop(columns ='Avg_Area_Number_of_Bedrooms', inplace=True)
```

**TRAIN TEST**

```python
m = np.random.rand(len(df))<0.8
train = df[m]
test = df[~m]
```

**LINEAR MODEL**

```python
from sklearn import linear_model

regr = linear_model.LinearRegression()

x_train = np.asanyarray(train[["Avg_Area_Income","Avg_Area_House_Age","Avg_Area_Number_of_Rooms","Area_Population"]])
y_train = np.asanyarray(train[["Area_Population"]])

regr.fit(x_train, y_train)

print("coef:" ,regr.coef_)
print("intercept:" ,regr.intercept_)
```

```python
x_test = np.asanyarray(test[["Avg_Area_Income","Avg_Area_House_Age","Avg_Area_Number_of_Rooms","Area_Population"]])
y_test = np.asanyarray(test[["Area_Population"]])

y_hat = regr.predict(x_test)
y_hat
```

**ACCURACY**

```python
from sklearn.metrics import r2_score

print("r2_score:",r2_score(y_test,y_hat))
```
